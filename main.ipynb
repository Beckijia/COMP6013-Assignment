#### Null Values

```python
# Define the function to check if there is any null value in each column of the dataframe
def missing(df):
  missing_dict = dict()
  total = df.shape[0]
  for attribute in df.columns:
    missing = df[attribute].isna().sum()
    frac = missing/total * 100
    missing_dict[attribute] = frac
  return missing_dict
```

```python
# Verify if there is any null value for the training data
train_missing_dict = missing(train_df_encode)
train_missing_dict  # The result shows all the training data has no null value.
```

```python
from sklearn.preprocessing import StandardScaler

#Processing both X_train and X_test using standarization
transformer = StandardScaler().fit(X_train_tts_nodummy)
X_train_tts_nodummy_normalized = transformer.transform(X_train_tts_nodummy)
X_test_tts_nodummy_normalized = transformer.transform(X_test_tts_nodummy)
```

#### Random Forest

##### Model Training and Hyperparameter Tuning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
# Choose the best parameters for Random Forest model

# Define the model
rf = RandomForestRegressor(random_state = 42)

# Set up hyperparameter
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [4, 6, 8],
    'min_samples_split': [2, 5, 8],
    'min_samples_leaf': [1, 2, 4]
}

# Setup the GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv = 5, verbose=2, n_jobs=-1)

# Fit the grid search by using normalized training data
grid_search.fit(X_train_tts_normalized, y_train_tts_normal)

# Search the best parameters by using the method of GridSearchCV.best_params_
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```

```python
# Get the model with the optimized parameters provided by GridSearchCV
model_rf = grid_search.best_estimator_
```

```python
from sklearn.metrics import mean_absolute_percentage_error

# Predict target value by using trained Random Forest model with the normalized training data
y_train_rf_pred = model_rf.predict(X_train_tts_normalized)
```

```python
# Predict target value by using X_test data with the trained Random Forest model
y_rf_pred = model_rf.predict(X_test_tts_normalized)
y_rf_pred
```
